```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def pd(qd):
    return 15-5*qd
def po(qo):
    return -14+5*qo
qd=np.linspace(0,5)
qo=np.linspace(0,5)
plt.plot(qd,pd(qd),'r',label="Demand")
plt.plot(qo,po(qo),'b',label="Supply")
plt.xlabel("quantity")
plt.ylabel("price")
plt.grid(True)
plt.legend()
plt.title("Supply and demand")
    
```
